```python
import requests
from bs4 import BeautifulSoup

# Función para extraer ingresos de Yahoo Finance
def get_revenue(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    income_statements = soup.find_all('div', {'data-test': 'fin-row'})
    revenue_data = []
    for statement in income_statements:
        if 'Total Revenue' in statement.text:
            revenue_text = statement.find_all('span')
            for text in revenue_text[1:]:  # Ignorar el primer elemento que contiene el texto 'Total Revenue'
                revenue_data.append(text.text)
            break
    return revenue_data

# URL de Yahoo Finance para Tesla
tesla_url = 'https://finance.yahoo.com/quote/TSLA/financials?p=TSLA'
tesla_revenue = get_revenue(tesla_url)

# Limpiar y estructurar los datos
def clean_revenue_data(revenue_data):
    clean_data = []
    for value in revenue_data:
        try:
            clean_data.append(int(value.replace(',', '').replace(' ', '')))
        except ValueError:
            continue
    return clean_data

tesla_revenue_cleaned = clean_revenue_data(tesla_revenue)
print("Tesla Revenue:", tesla_revenue_cleaned)
```

```python
import yfinance as yf
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

# Descargar datos de acciones de Tesla
tesla_stock_data = yf.download('TSLA', start='2020-01-01', end='2023-01-01')

# Función para extraer ingresos de Yahoo Finance
def get_revenue(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    income_statements = soup.find_all('div', {'data-test': 'fin-row'})
    revenue_data = []
    for statement in income_statements:
        if 'Total Revenue' in statement.text:
            revenue_text = statement.find_all('span')
            for text in revenue_text[1:]:
                revenue_data.append(text.text)
            break
    return revenue_data

# URL de Yahoo Finance para Tesla
tesla_url = 'https://finance.yahoo.com/quote/TSLA/financials?p=TSLA'
tesla_revenue = get_revenue(tesla_url)

# Limpiar y estructurar los datos de ingresos
def clean_revenue_data(revenue_data):
    clean_data = []
    for value in revenue_data:
        try:
            clean_data.append(int(value.replace(',', '').replace(' ', '')))
        except ValueError:
            continue
    return clean_data

tesla_revenue_cleaned = clean_revenue_data(tesla_revenue)

# Verificar los datos extraídos y limpiados
print("Tesla Revenue Raw Data:", tesla_revenue)
print("Tesla Revenue Cleaned Data:", tesla_revenue_cleaned)

# Graficar datos de acciones de Tesla
tesla_stock_data['Close'].plot(title='Tesla Stock Prices', figsize=(10, 6))
plt.show()

# Graficar ingresos de Tesla
years = list(range(2020, 2020 + len(tesla_revenue_cleaned)))
plt.figure(figsize=(10, 6))
plt.plot(years, tesla_revenue_cleaned, label='Tesla Revenue')
plt.xlabel('Year')
plt.ylabel('Total Revenue (in billions)')
plt.title('Tesla Revenue')
plt.legend()
plt.show()
```
